# **ชุดข้อมูลที่ 2 ปริมาณของกลางยาเสพติด**

```python
data_02 = pd.read_csv(os.path.join(path,'ปริมาณของกลางยาเสพติด.csv')) # ไฟล์ที่เราต้องการ
data_02 # print data ดูว่ามันหน้าตาเป็นยังไง
```

### **คำอธิบายตาราง**

*   **รหัสจังหวัด** คือ รหัสของจังหวัดที่พบของกลางยาเสพติด
*   **จังหวัด** คือ ชื่อของจังหวัดที่พบของกลางยาเสพติด
*   **ปริมาณของกลางยาเสพติดทั้งหมด (กิโลกรัม)** คือ ปริมาณของกลางยาเสพติดที่พบในแต่ละจังหวัดทั้งหมด (กิโลกรัม)
*   **ชนิดของกลางยาเสพติด** คือ ชนิดของยาเสพติดที่พบในแต่ละจังหวัด
*   **ยาบ้า (กิโลกรัม)** คือ ปริมาณของยาบ้าที่พบในแต่ละจังหวัด (กิโลกรัม)
*   **ไอซ์ (กิโลกรัม)** คือ ปริมาณของยาไอซ์ที่พบในแต่ละจังหวัด (กิโลกรัม)
*   **กัญชาแห้ง (กิโลกรัม)** คือ ปริมาณของกัญชาแห้งที่พบในแต่ละจังหวัด (กิโลกรัม)
*   **เฮโรอีน (กิโลกรัม)** คือ ปริมาณของเฮโรอีนที่พบในแต่ละจังหวัด (กิโลกรัม)
*   **คีตามีน (กิโลกรัม)** คือ ปริมาณของคีตามีนที่พบในแต่ละจังหวัด (กิโลกรัม)
*   **โคเคน (กิโลกรัม)** คือ ปริมาณของโคเคนที่พบในแต่ละจังหวัด (กิโลกรัม)
*   **กระท่อม (กิโลกรัม)** คือ ปริมาณของกระท่อมที่พบในแต่ละจังหวัด (กิโลกรัม)



```python
data_02 = data_02.sort_values(by=['รหัสจังหวัด']) # เรียงข้อมูลตามรหัสจังหวัด 
data_02
```

### **ตรวจสอบค่า Missing ของตาราง data_02**

```python
data_02.isnull() # ตรงไหนเป็นช่องว่างจะขึ้นว่า True ตรงไหนไม่เป็นช่องว่างจะขึ้นว่า False
```

```python
data_02.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง สรุปคือทุกคอลัมน์ไม่มีค่า  missing 
```

```python
data_02_cut = data_02.iloc[:,4:] # เลือกเฉพาะคอลัมน์ที่เป็นจำนวนของปริมาณยาเสพติด (กิโลกรัม)
data_02_cut
```

```python
data_02_value = data_02_cut.values # แปลงจากตารางให้กลายเป็นค่าของข้อมูล
data_02_value
```

```python
import numpy as np
max_all = np.zeros(data_02_value.shape[0]) # หาค่าที่มากที่สุดของแต่ละแถวเพื่อดูว่ายาเสพติดชนิดที่พบมากที่สุดมีปริมาณเท่าไหร่
for row in range(data_02_value.shape[0]):
    max_all[row] = max(data_02_value[row,:])
print(max_all)
```

```python
max_all = pd.DataFrame(max_all) # ดูว่าปริมาณของยาเสพติดที่พบมากที่สุดของแต่ละแถวหรือแต่ละจังหวัดเป็นเท่าไหร่
print(max_all)
```

```python
data_02 = data_02.assign(max_all = max_all.values) # เพิ่มคอลัมน์ของ max_all เข้าไปยังตารางข้อมูล
data_02
```

```python
data_02.rename(columns={'max_all':'ปริมาณของกลางยาเสพติดที่พบมากที่สุด (กิโลกรัม)'}, inplace=True) # เปลี่ยนชื่อคอลัมน์
data_02
```

```python
data_02.sort_values(by='ปริมาณของกลางยาเสพติดที่พบมากที่สุด (กิโลกรัม)') # เรียงข้อมูลตามปริมาณของกลางยาเสพติดที่พบมากที่สุด (กิโลกรัม)
```

```python
# ดูว่ามียาเสพติดชนิดใดบ้างที่พบปริมาณมากที่สุด
data_02.groupby(['ชนิดของกลางยาเสพติด']).sum().sort_values(by='ปริมาณของกลางยาเสพติดที่พบมากที่สุด (กิโลกรัม)')
```

```python
data_02['ชนิดของกลางยาเสพติด'].replace({  # ตรงคอลัมน์ ชนิดของกลางยาเสพติด ลบหน่วยคำว่า (กิโลกรัม) ออกไปให้เหลือแค่ชื่อของชนิดยาเสพติดเพียงอย่างเดียว
        'กัญชาแห้ง (กิโลกรัม)':'กัญชาแห้ง',
        'ไอซ์ (กิโลกรัม)':'ไอซ์',
        'ยาบ้า (กิโลกรัม)':'ยาบ้า',
        'กระท่อม (กิโลกรัม)':'กระท่อม'
    }, inplace=True)
data_02
```

```python
# ดูว่ามียาเสพติดชนิดใดบ้างที่พบปริมาณมากที่สุด
data_02.groupby(['ชนิดของกลางยาเสพติด']).sum().sort_values(by='ปริมาณของกลางยาเสพติดที่พบมากที่สุด (กิโลกรัม)')
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix  # เรียกใช้อัลกอริทึมที่เราต้องการ
```

เลือกตัวแบบ **KNN 2** แล้วนำมา Train ใหม่

```python
# Define
neigh2_final = KNeighborsClassifier(n_neighbors=10, weights='distance') # ถามเพื่อนบ้าน 2 คน และเชื่อเพื่อนบ้านที่ใกล้ที่สุดมากกว่าเพื่อนบ้านที่ไกล
#Train
neigh2_final.fit(X_train,y_train)
#Test
KNN2final_result = neigh2_final.predict(X_test)
accuracy_score(y_test, KNN2final_result) # นั่นคือ มีความแม่นยำ 35.90 %
```

เลือกตัวแบบ **KNN 3** แล้วนำมา Train ใหม่

```python
# Define
neigh3_final = KNeighborsClassifier(n_neighbors=1) # เลือกเพื่อนบ้านที่ใกล้ที่สุดแค่คนเดียว
#Train
neigh3_final.fit(X_train,y_train)
#Test
KNN3final_result = neigh3_final.predict(X_test)
accuracy_score(y_test, KNN3final_result) # นั่นคือ มีความแม่นยำ 35.90 %
```

เลือกตัวแบบ **ANN 1** แล้วนำมา Train ใหม่

```python
#Define
perceptron1_final = MLPClassifier(random_state=6, max_iter=2000,hidden_layer_sizes=1,learning_rate_init=0.005)  # กำหนดพารามิเตอร์ให้กับ model
#Train
perceptron1_final.fit(X_train,y_train) 
#Test
perceptron1final_result = perceptron1_final.predict(X_test)
accuracy_score(y_test,perceptron1final_result) # นั่นคือ มีความแม่นยำ 30.77 %
```

เลือกตัวแบบ **ANN 2** แล้วนำมา Train ใหม่

```python
#Define
perceptron2_final = MLPClassifier(random_state=6, max_iter=2000,hidden_layer_sizes=10,learning_rate_init=0.005)
#Train
perceptron2_final.fit(X_train,y_train) 
#Test
perceptron2final_result = perceptron2_final.predict(X_test)
accuracy_score(y_test,perceptron2final_result) # นั่นคือ มีความแม่นยำ 71.79 %
```

ดังนั้น เมื่อนำทั้ง 4 ค่ามาเทียบกันพบว่า ค่า accuracy ของ ANN 2 มีค่ามากที่สุด นั่นคือมีความแม่นยำ 71.79 % จึงเลือกตัวแบบ ANN 2 เพราะมีความเหมาะสมและมีความแม่นยำที่สุด
