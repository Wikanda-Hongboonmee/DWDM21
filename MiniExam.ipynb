## **2. วาด Histogram แสดงจำนวนเครื่องบินขาออกของแต่ละวันในสัปดาห์**

````python
merged_data = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน') 
merged_data
# how คือจะเอาตารางไหนเป็นตารางหลัก ในที่นี้เอาตาราง flights เป็นตารางหลัก
# left_on จะเอา index ของตารางซ้าย ( flights ) อันไหนมาผูกกับตารางนี้
# right_on จะเอา index ของตารางขวา ( airports ) อันไหนมาผูกกับตารางนี้
````

````python
X1 = merged_data.groupby('วันในสัปดาห์').count() # count() คือ นับจำนวนเครื่องบินขาออกและทำการจัดกลุ่ม หรือ Groupby 'วันในสัปดาห์' เพื่อดูจำนวนเครื่องบินขาออกของแต่ละวันในสัปดาห์
X1
````

````python
X1_Select = X1['รหัสสนามบินต้นทาง'] # เลือกคอลัมน์ 'รหัสสนามบินต้นทาง' เพื่อดูจำนวนเครื่องบินขาออกของแต่ละวันในสัปดาห์ 
X1_Select 
````

````python
from matplotlib import pyplot as plt # import package เพื่อนำมาใช้ในการ plot กราฟ histogram
````

````python
# the histogram of the data
plt.hist(X1_Select,4, facecolor='r', alpha=1) # กำหนด facecolor='r' ก็คือกำหนดสี ในที่นี้กำหนดให้เป็นสีแดง (red)

plt.xlabel('quantity of departures') # กำหนด label ของแกน x 
plt.ylabel('day of the week') # กำหนด label ของแกน y
plt.title('quantity of departures in the day of the week') # กำหนดหัวข้อหรือ title ของกราฟ histogram

plt.grid(True)
plt.show()
````

## **3. สายการบินไหนมีเที่ยวบินมากที่สุดและน้อยที่สุด**

````python
X2 = merged_data.groupby('สายการบิน').count() # count() คือ นับจำนวนเที่ยวบินและทำการจัดกลุ่ม หรือ Groupby 'สายการบิน' เพื่อดูจำนวนเที่ยวบินของแต่ละสายการบิน
X2
````

````python
X2.sort_values(by=['วันในเดือน'],ascending=False) # sort_values คือ การเรียงลำดับข้อมูล โดยใช้คอลัมน์ 'วันในเดือน' เพื่อดูจำนวนเที่ยวบินของแต่ละสายการบินจากมากไปน้อย กำหนดได้โดย ascending=False
                                                                        # เนื่องจากหากจำนวนของวันในเดือนมาก แสดงว่าจำนวนเที่ยวบินก็จะมากไปด้วย 
````

````python
X2.sort_values(by=['วันในเดือน'],ascending=False).iloc[:1,:] # เลือกมาเฉพาะแถวแรกสุด นั่นคือ สายการบิน WN มีจำนวนเที่ยวบินมากที่สุดคือ 575739 วันในเดือน 
````

````python
X2.sort_values(by=['วันในเดือน'],ascending=False).iloc[-1:,:] # เลือกมาเฉพาะแถวสุดท้าย นั่นคือ สายการบิน HA มีจำนวนเที่ยวบินน้อยที่สุดคือ 17432 วันในเดือน 
````

- วันไหนในสัปดาห์ที่มีจำนวนเครื่องบินขาเข้ามากที่สุด 3 อันดับแรก



````python
X3 = merged_data.groupby('วันในสัปดาห์').count() # count() คือ จำนวนเครื่องบินขาเข้าและทำการจัดกลุ่ม หรือ Groupby 'วันในสัปดาห์' เพื่อดูจำนวนเครื่องบินขาเข้าของแต่ละวันในสัปดาห์
X3
````

````python
X3.sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) # sort_values คือ การเรียงลำดับข้อมูล โดยใช้คอลัมน์ 'รหัสสนามบินปลายทาง' เพื่อดูจำนวนเครื่องบินขาเข้าจากมากไปน้อย กำหนดได้โดย ascending=False
                                                                                        # เลือกคอลัมน์ 'รหัสสนามบินต้นทาง' เพื่อดูจำนวนเครื่องบินขาเข้าของแต่ละวันในสัปดาห์ 
````

````python
X3.sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False).iloc[:3,:] # เลือกมาเฉพาะ 3 แถวแรก นั่นคือ วันจันทร์มีจำนวนเครื่องบินขาเข้ามากที่สุด คือ 407837 รองลงมา คือ วันพฤหัสมีจำนวนเครื่องบินขาเข้า คือ 406563
                                                                                                    # และอันดับที่ 3 คือวันพุธมีจำนวนเครื่องบินขาเข้า คือ 403072
````
